```python
import pandas as pd
```

```python
citi202005_data = pd.read_csv("Raw_Data/JC-202005-citibike-tripdata.csv")
```

```python
citi202005_data
```

```python
#Extract the date of the start and end time
citi202005_data['start_date'] = pd.to_datetime(citi202005_data['starttime'])
citi202005_data['start_date'] = citi202005_data['start_date'].dt.date

citi202005_data['end_date'] = pd.to_datetime(citi202005_data['stoptime'])
citi202005_data['end_date'] = citi202005_data['end_date'].dt.date

citi202005_data
```

```python
#Add Age field given the current year is 2020
citi202005_data["2020_Age"] = 2020 - citi202005_data["birth year"]
citi202005_data
```

```python
##Trip duration isn't right, need to correct this!
citi202005_data["Duration(mins)"] = citi202005_data["tripduration"]/60
citi202005_data["Duration(hrs)"] = citi202005_data["Duration(mins)"]/60
citi202005_data
```
